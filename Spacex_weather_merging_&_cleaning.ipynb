# Function for Fetch relative Weather Data

```python
import requests

def get_weather_openmeteo(lat,lon,date):
    url_2 = (
        f"https://archive-api.open-meteo.com/v1/archive?"
        f"latitude={lat}&longitude={lon}"
        f"&start_date={date}&end_date={date}"
        f"&hourly=temperature_2m,relative_humidity_2m,weathercode,wind_speed_10m"
        f"&timezone=UTC"
    )


    try:
        response = requests.get(url_2)
        if response.status_code == 200:
            return response.json()
        else:
            return None
    except Exception as e:
        print("error: ",e)
        return None
        
        
```

```python
# launchpad ID & date of every luanch
launches = data[['date_utc', 'launchpad', 'name']].copy()

#convert date in YY-MM-DD method
launches['date'] = pd.to_datetime(launches.date_utc).dt.date

# fetch lat/lon from Launchpad ID
launches['lat'] = launches.launchpad.map(lambda x: launchpad_location.get(x, (None, None))[0])
launches['lon'] = launches.launchpad.map(lambda x: launchpad_location.get(x, (None, None))[1])

# keep only required columns
launches = launches[['name', 'date', 'lat', 'lon']]
launches
```

# Relative Weather Data Fetching

```python
# fetch weather data by using function get_weather_openmeteo
weather_data = []

for _, row in launches.iterrows():
    weather = get_weather_openmeteo(row.lat, row.lon, row.date)
    weather_data.append(weather)

weather_data
```

# Data Cleaning

```python
weather_df = pd.DataFrame()

for i, entry in enumerate(weather_data):
    if entry and 'hourly' in entry:
        df = pd.DataFrame(entry['hourly'])
        df['launch_index'] = i
        weather_df = pd.concat([weather_df, df], ignore_index=True)
```

```python
# merging launch and weather data

launches['date_utc'] = data.date_utc
launches['launch_hour'] = pd.to_datetime(launches['date_utc']).dt.floor('h')

weather_df.time = pd.to_datetime(weather_df.time, utc=True)
weather_df["launch_hour"] = weather_df.time.dt.floor("h")
weather_df = weather_df.drop_duplicates(subset="launch_hour")

merged_weather = pd.merge(launches, weather_df, on= "launch_hour", how="left")
```
